### Time-Series Components

```python
date_cols = [
    col
    for col in df_retail.columns
    if "date" in col.lower() or "timestamp" in col.lower()
]
```

```python
# Convert relevant columns to datetime
# Note: Error handling for parsing will be more robust in Task 1.6
df_retail["PurchaseDate"] = pd.to_datetime(df_retail["PurchaseDate"], errors="coerce")
df_retail["SignUpDate"] = pd.to_datetime(df_retail["SignUpDate"], errors="coerce")
df_retail["LastLogin"] = pd.to_datetime(df_retail["LastLogin"], errors="coerce")
```

```python
print("Date/Time Columns converted to datetime type\n")
print(df_retail[["PurchaseDate", "SignUpDate", "LastLogin"]].info())
```

```python
# Aggregate total amount by purchase date
daily_sales = (
    df_retail.groupby(df_retail["PurchaseDate"].dt.date)["TotalAmount"]
    .sum()
    .reset_index()
)
daily_sales["PurchaseDate"] = pd.to_datetime(daily_sales["PurchaseDate"])
daily_sales.set_index("PurchaseDate", inplace=True)

# Resample to weekly and monthly for clearer trends
weekly_sales = daily_sales["TotalAmount"].resample("W").sum()
monthly_sales = daily_sales["TotalAmount"].resample("ME").sum()
```

```python
print("Daily Sales Head:")
display(daily_sales.head())
print("Weekly Sales Head:")
display(weekly_sales.head())
print("Monthly Sales Head:")
display(monthly_sales.head())
```

```python
# Visualizing Sales Trends
plt.figure(figsize=(15, 7))
plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker="o",
    linestyle="-",
    color="skyblue",
)
plt.title("Monthly Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales Amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

from March 2021 up to aproximately December 2021 total sales looks stable, ranging at $10 Milion. But, in end of month sales, there is decrease that show up, sales decrease from **$10 Milion** to under **$1 Milion**, from the source of the data, **this data is cut by the owner**

```python
plt.figure(figsize=(15, 7))
plt.plot(
    weekly_sales.index,
    weekly_sales.values,
    marker="o",
    linestyle="-",
    color="lightcoral",
)
plt.title("Weekly Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales Amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

There was a significant spike in the early weeks of January 2021, with sales increasing dramatically from around **$1 million** to around **$2.4 million**. As of January 2021, weekly sales are expected to stabilize at around **$2.3 million** to **$2.5 million** from around the end of January 2021 to December 2021. Similar to the quarterly chart, there is a clear decline in sales in the last weeks shown (around late December 2021/early January 2022). the reason behind it, because **data cut by owner**

```python
# Daily/Monthly Active Users (based on LastLogin)

# Daily Active Users (DAU)
# Count unique users per day based on LastLogin
dau = (
    df_retail.groupby(df_retail["LastLogin"].dt.date)["UserID"].nunique().reset_index()
)
dau.columns = ["Date", "ActiveUsers"]
dau["Date"] = pd.to_datetime(dau["Date"])
dau.set_index("Date", inplace=True)

# Monthly Active Users (MAU)
# For MAU, it's unique users logging in within each month period.
# Extract year and month, then count unique users
df_retail["LastLogin_Month"] = df_retail["LastLogin"].dt.to_period("M")
mau = df_retail.groupby("LastLogin_Month")["UserID"].nunique().reset_index()
mau.columns = ["Month", "ActiveUsers"]
mau["Month"] = mau[
    "Month"
].dt.to_timestamp()  # Convert Period to Timestamp for plotting
mau.set_index("Month", inplace=True)
```

```python
print("Daily Active Users Head:")
display(dau.head())
print("Monthly Active Users Head:")
display(mau.head())
```

```python
# Visualizing Active Users Trends
plt.figure(figsize=(15, 6))
plt.plot(mau.index, mau.values, marker="o", linestyle="-", color="mediumseagreen")
plt.title("Monthly Active Users (MAU) Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Unique Users")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**MAU** data shows relatively stable user activity throughout most of 2024, with reasonable monthly fluctuations, ranging between **7,900** and **8,500** unique users per month. The sharp decline at the end of the period can be disregarded as a business anomaly as it was confirmed to be the result of data truncation.

```python
plt.figure(figsize=(15, 6))
plt.plot(dau.index, dau.values, marker=".", linestyle="-", color="slateblue", alpha=0.7)
plt.title("Daily Active Users (DAU) Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Unique Users")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Unlike **MAU**, which is smoother, **DAU** shows significant daily fluctuations. This is normal in daily activity data, where the number of users can vary from day to day due to usage patterns (e.g., weekends vs. weekdays, holidays, etc.). Despite these fluctuations, the majority of **DAU** is in the range of **260-300** unique users per day. There are a few peaks above **300** and valleys below **240**.

## Insight


- High Data Quality (Minimal Missing Values): The `df_retail` dataset is surprisingly clean, with no missing values in either the numeric or categorical columns.

- Uniform Distribution of Numeric Features: Most numeric features such as `UserID`, `ProductID`, `Age`, `Price`, `Quantity`, and `DiscountRate` show a relatively uniform distribution.

    - `Quantity` is a discrete variable with values of 1, 2, 3, and 4.

- Confirmed `ReviewScore` Anomaly: The `ReviewScore` column shows values outside the range of **1−5** (minimum **−0.60** and maximum **8.60**), which is an anomaly and needs to be addressed during the cleaning stage. The presence of `ReviewText` with discrete sentiment (`Excellent`, `Good`, `Poor`, `Average`) further reinforces the need to recalibrate `ReviewScore`.

- `TotalAmount` Right-Skewed: The `TotalAmount` column has a right-skewed distribution (positively skewed) with outliers at very high transaction values. This is normal for sales data and may require logarithmic transformation in the future.

- Logical Correlation: `TotalAmount` is highly positively correlated with `Price` (**0.74**) and `Quantity` (**0.58**), which makes business sense. Other numerical features show very weak linear correlations.

- Balanced Categorical Features: All categorical features such as `Gender`, `Country`, `ProductName`, `Category`, `ReviewText`, `DeviceType`, and `ReferralSource` show very balanced distributions across their categories.

- Time Trends:

- Monthly and weekly sales show stable trends after an initial ramp-up in 2021.

    - Monthly active users (MAU) also show stability in 2024.

    - The sharp decline in sales and MAU at the end of the period is confirmed as a result of data cuts by the owner, not a business anomaly.

- Daily active users (DAU) show normal daily fluctuations but no clear overall upward/downward trend.
